# Coronary Heart Disease Analysis and Data Visualization

### Introduction

In this Jupyter Notebook, we perform an analysis of coronary heart disease (CHD) data and create data visualizations to gain insights into the dataset. The analysis includes the following steps:

- Importing the required packages
- Importing and displaying dataset information
- Analyzing numerical features
- Analyzing the target variable
- Bivariate analysis to explore relationships between features

### 1. Importing the Required Packages

Let's start by importing the necessary Python packages for our analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Importing & Displaying Dataset Information

```python
df = pd.read_csv('data/Suivi.csv')
print("Data's shape : ", df.shape)
df.head()
```

### 3. Analyzing Numerical Features

We analyze numerical features in the dataset, including their data types and summary statistics.

```python
df.info()
df.describe()
```

- Sex (Sex): Indicates the gender of the patient. Values: Male (0) or Female (1). Nominal variable.

- Age (Age): Represents the age of the patient. Although the recorded ages are rounded to whole numbers, the concept of age is continuous.

- Current Smoker (isSmoker): Indicates whether the patient is currently a smoker. Values: Yes or No. Nominal variable.

- Cigarettes Per Day (CigsPerDay): Reflects the average number of cigarettes smoked by the person per day. While discrete in count, it can be treated as continuous due to the range of possible values, including fractional amounts like half a cigarette.

- Blood Pressure Medication (BloodPressureMeds): Indicates whether the patient was taking blood pressure medication. Values: Yes or No. Nominal variable.

- Prevalent Stroke (PrevStroke): Indicates whether the patient has a history of stroke. Values: Yes or No. Nominal variable.

- Prevalent Hypertension (PrevHypertension): Indicates whether the patient was diagnosed with hypertension. Values: Yes or No. Nominal variable.

- Diabetes (Diabetes): Indicates whether the patient has diabetes. Values: Yes or No. Nominal variable.

- Total Cholesterol (Cholesterol): Represents the total cholesterol level. Continuous variable.

- Systolic Blood Pressure (SysBloodPressure): Represents the systolic blood pressure. Continuous variable.

- Diastolic Blood Pressure (DiaBloodPressure): Represents the diastolic blood pressure. Continuous variable.

- Body Mass Index (BMI): Represents the Body Mass Index, which is a continuous measure of body composition.

- Heart Rate (HeartRate): Represents the heart rate. Although technically discrete, it is considered continuous due to the large number of possible values.

- Glucose Level (Glucose): Represents the glucose level. Continuous variable.

- 10-Year Risk of Coronary Heart Disease (RiskCHD): This is the target variable. It's binary, where "1" indicates a presence of risk and "0" indicates no risk. Nominal variable.

These attributes provide comprehensive information about the patients' health and lifestyle factors, contributing to the assessment of their 10-year risk of coronary heart disease.

##### Calculate the correlation matrix

```python
correlation_matrix = df.corr()

# Set up the plot
plt.figure(figsize=(12, 8))

# Create a heatmap of the correlation matrix
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(heatmap, fraction=0.046, pad=0.04)  # Add a colorbar for reference

# Set x and y tick labels
plt.xticks(np.arange(len(correlation_matrix)), correlation_matrix.columns, rotation=45)
plt.yticks(np.arange(len(correlation_matrix)), correlation_matrix.columns)

# Set plot title and labels
plt.title("Correlation Heatmap", size=18)
# Show the plot
plt.show()
correlation_matrix
```

We notice that only the attributes 'Age' and 'SysBloodPressure' have the highest correlation coefficients in relation to the RiskCHD attribute.

```python
numeric_features = ['CigsPerDay', 'Cholesterol', 'SysBloodPressure', 'DiaBloodPressure', 'BMI', 'HeartRate', 'Glucose']

# Calculate the number of rows and columns needed for the subplot grid
num_rows = (len(numeric_features) - 1) // 2 + 1
num_cols = 2

# Set up the main figure for subplots
plt.figure(figsize=(18, 5 * num_rows), facecolor='w')

# Loop through each numerical feature and create a subplot
for idx, feature in enumerate(numeric_features, start=1):
    plt.subplot(num_rows, num_cols, idx)  # Create a subplot
    sns.histplot(df[feature], kde=True, color='skyblue')  # Plot the histogram with KDE
    plt.title('{} Distribution'.format(feature), fontsize=16)  # Set subplot title
    plt.xlabel(feature)  # Label the x-axis with the feature name
    plt.ylabel('Frequency')  # Label the y-axis
    
# Adjust the layout for better spacing between subplots
plt.tight_layout()

# Display the subplots
plt.show()
```

We observe a significant imbalance in the distribution of the 'CigsPerDay' feature, with a substantial portion of the data concentrated at 0. In contrast, other features such as 'Cholesterol', 'SysBloodPressure', 'DiaBloodPressure', and 'BMI' exhibit more uniform distributions.

### 4. Analyzing the Target Variable

We visualize the distribution of the target variable 'RiskCHD' using a pie chart to understand the class distribution.

```python
plt.pie(df['RiskCHD'].value_counts(), labels=['0', '1'], colors = ['lightcoral', 'lightgreen'], shadow=True, explode=(0.2, 0), startangle=90, autopct='%1.1f%%')
plt.title("Percentage of Risk CHD", fontsize=12, fontweight='bold')
plt.show()
```

### 5. Bivariate Analysis: Exploring Relationships Between Features

#### 'Sex' and 'RiskCHD' : Determining which gender demonstrates a greater susceptibility to coronary heart disease (CHD)

```python
# Filter the DataFrame for RiskCHD = 1
df_chd_positive = df[df['RiskCHD'] == 1]

plt.pie(df_chd_positive['Sex'].value_counts(), labels=['Male', 'Female'], colors = ['skyblue', 'red'], shadow=True, explode=(0.1, 0), startangle=90, autopct='%1.1f%%')
plt.title("Percentage of Risk CHD by Gender", fontsize=12, fontweight='bold')
plt.show()
```

#### 'CigsPerDay' and 'RiskCHD' : Investigating the Link Between Daily Cigarette Consumption and the Risk of Coronary Heart Disease

```python
plt.figure(figsize=(14, 8))

# Create count plot with hue
ax = sns.countplot(x="RiskCHD", data=df, hue="CigsPerDay", palette='coolwarm')

# Adjust legend size and position
legend = ax.legend(title='CigsPerDay', fontsize='x-small', loc='upper right')

# Set plot title and axis labels
plt.title("Relation between CigsPerDay and Risk of Coronary Heart Disease", size=18)
plt.xlabel("Risk of CHD", size=14)
plt.ylabel("Count", size=14)
plt.xticks(size=12)
plt.yticks(size=12)

# Adjust layout to accommodate the legend
plt.subplots_adjust(right=0.85)

plt.show()
```

We observe that lower 'CigsPerDay' is tied to reduced CHD risk : Non-smokers ('CigsPerDay' = 0.0) exhibit notably lower disease risk. 
However, this correlation doesn't ensure substantially lower CHD risk.

#### Blood Pressure and CHD Risk

We visualize the relationship between 'SysBloodPressure', 'DiaBloodPressure', and CHD risk.

```python
colors = {False: 'blue', True: 'red'}
labels = {False: '0', True: '1'}

plt.scatter(df[df['RiskCHD'] == False]['SysBloodPressure'], df[df['RiskCHD'] == False]['DiaBloodPressure'],
            c=colors[False], label=labels[False], alpha=0.5)
plt.scatter(df[df['RiskCHD'] == True]['SysBloodPressure'], df[df['RiskCHD'] == True]['DiaBloodPressure'],
            c=colors[True], label=labels[True], alpha=0.5)

plt.xlabel('SysBloodPressure')
plt.ylabel('DiaBloodPressure')
plt.legend()
plt.show()
```

##### Cholesterol, Age, and CHD Risk

```python
colors = {False: 'green', True: 'purple'}
labels = {False: '0', True: '1'}

plt.scatter(df[df['RiskCHD'] == False]['Age'], df[df['RiskCHD'] == False]['Cholesterol'],
            c=colors[False], label=labels[False], alpha=0.5)
plt.scatter(df[df['RiskCHD'] == True]['Age'], df[df['RiskCHD'] == True]['Cholesterol'],
            c=colors[True], label=labels[True], alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.show()
```

These visualizations help us understand the relationships between various features and the risk of coronary heart disease, providing valuable insights for further analysis and decision-making.
